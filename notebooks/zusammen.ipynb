```python
import pickle
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
from scipy.stats import t


import pandas as pd

model_names = [
    "NaiveSA",                # ok   
    "AFinn",                  # ok
    "Vader", 
    "RandomForestW2V",        # ko
    "RandomForest",           # ko
    "LogisticRegression",     # ok
    "LogisticRegressionW2V",  # ok
    "LSTM",
    "LSTM_W2V"
]

def load_data(output_dir, model_name):
    file_name = os.path.join(output_dir, 'grid_search_{}_cv_results.pkl'.format(model_name))
    with open(file_name, 'rb') as f:
        results = pickle.load(f)
    results_df = pd.DataFrame(results)
    # results_df.columns
    return results_df

data = {}
labels = {}
for metric in ["accuracy", "f1", "recall", "precision"]:

    data[metric] = []
    labels[metric] = []
    for model_name in model_names:
        try:
            output_dir = "/home/zhanna/bachelorarbeit/zbb/experiments/output_grid_search_000"
            df = pd.DataFrame(load_data(output_dir, model_name))
            df = df.sort_values(by=["rank_test_{}".format(metric)])
            data[metric].append(df['mean_test_{}'.format(metric)].iloc[0])
            labels[metric].append(model_name)
        except Exception as e:
            print(" cannot load file for model ", model_name, ": ", str(e))
        print(" model", model_name)

colors = ['red', 'green', 'blue', 'magenta']
for metric, c in zip(data.keys(), colors):
    fig, ax = plt.subplots()
    ax.bar(labels[metric], data[metric], color=c)
    ax.set_xticks([i for i in range(0, len(labels[metric]))], labels[metric], rotation=90)
    ax.grid('minor')
    ax.set_title("Vergleich der Metrik '{}' zwischen Modellen".format(metric))
```
